# LAB | Imbalanced

**Load the data**

In this challenge, we will be working with Credit Card Fraud dataset.

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/card_transdata.csv

Metadata

- **distance_from_home:** the distance from home where the transaction happened.
- **distance_from_last_transaction:** the distance from last transaction happened.
- **ratio_to_median_purchase_price:** Ratio of purchased price transaction to median purchase price.
- **repeat_retailer:** Is the transaction happened from same retailer.
- **used_chip:** Is the transaction through chip (credit card).
- **used_pin_number:** Is the transaction happened by using PIN number.
- **online_order:** Is the transaction an online order.
- **fraud:** Is the transaction fraudulent. **0=legit** -  **1=fraud**


```python
#Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import precision_score, recall_score, classification_report, confusion_matrix, f1_score
from sklearn.preprocessing import StandardScaler

from sklearn.utils import resample
```

```python
fraud = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/card_transdata.csv")
fraud.head()
```

```python
fraud.nunique()
```

```python
fraud.isnull().sum()
```

```python
#1. What is the distribution of our target variable? Can we say we're dealing with an imbalanced dataset?
```

```python
sns.countplot(data=fraud, x='fraud')
plt.title('Distribution of fraud')
plt.show()
```

We are indeed working with an imbalanced dataset, the number of no fraud transaction is higher then the fraud ones.

```python
# Train - Split 
```

```python
cont_cols= ['distance_from_home','distance_from_last_transaction','ratio_to_median_purchase_price']
cat_cols=['repeat_retailer','used_chip','used_pin_number','online_order','fraud']
target=['fraud']
```

```python
features = fraud.drop(columns = ["fraud"])
target = fraud["fraud"]

X_train, X_test, y_train, y_test = train_test_split(features, target,random_state=0)
```

```python
#scaling all data
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled_np = scaler.transform(X_train)
X_test_scaled_np = scaler.transform(X_test)
```

```python
#2. Train a LogisticRegression.
```

```python
log_reg = LogisticRegression()
X_train_scaled_df = pd.DataFrame(X_train_scaled_np, columns=X_train.columns, index=X_train.index)
X_test_scaled_df  = pd.DataFrame(X_test_scaled_np, columns=X_test.columns, index=X_test.index)
log_reg.fit(X_train_scaled_df, y_train)
```

```python
log_reg.score(X_test_scaled_df, y_test)
```

```python
#Evaluate your model. Take in consideration class importance, and evaluate it by selection the correct metric
```

```python
y_pred_test_log = log_reg.predict(X_test_scaled_df)
print(classification_report(y_pred = y_pred_test_log, y_true = y_test))
```

```python
#As it is important to  minimizes missed fraud, we are interested in the recall metric, catching as many frauds as possible. 
# Based on this metric the model is on point around 60% of the times, which indicates a weak model.
```

```python
#4. Run Oversample in order to balance our target variable and repeat the steps above, now with balanced data. 
#Does it improve the performance of our model?
```

```python
#Oversampling
from imblearn.over_sampling import RandomOverSampler

over = RandomOverSampler(random_state=0)
X_train_over, y_train_over = over.fit_resample(X_train_scaled_df,y_train)

#Training the model to see if any improvements
log_reg1 = LogisticRegression(max_iter=1000)
log_reg1.fit(X_train_over, y_train_over)
```

```python
#Evaluating the model
y_pred_test_os = log_reg1.predict(X_test_scaled_df)
print(classification_report(y_pred = y_pred_test_os, y_true = y_test))
```

```python
#now with balanced data using oversampling, we check Does it improve the performance of our model? and the answer is it does improve,
# and improves significantly as it went from 60% up to 95% being on point to predict a fraud as fraud.
```

```python
#5. Now, run Undersample in order to balance our target variable and repeat the steps above (1-3), 
#now with balanced data. Does it improve the performance of our model?
```

```python
#Undersampling
from imblearn.under_sampling import RandomUnderSampler

under = RandomUnderSampler(random_state=0)
X_train_under, y_train_under = over.fit_resample(X_train_scaled_df,y_train)

#Training the model to see if any improvements
log_reg2 = LogisticRegression(max_iter=1000)
log_reg2.fit(X_train_under, y_train_under)
```

```python
#Evaluating the model
y_pred_test_us= log_reg2.predict(X_test_scaled_df)
print(classification_report(y_pred = y_pred_test_us, y_true = y_test))
```

```python
#now with balanced data using undersampling, we check Does it improve the performance of our model? and the answer is it does improve
# compared to then unbalanced data, it goes from 60% up to 95% same as the score obtained iwth oversampling.
```

```python
#6. Finally, run SMOTE in order to balance our target variable and repeat the steps above (1-3), now with balanced data. 
#Does it improve the performance of our model?
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 1,sampling_strategy=1.0)
X_train_sm, y_train_sm =  sm.fit_resample(X_train_scaled_df,y_train) 

#Training the model to see if any improvements
log_reg3 = LogisticRegression(max_iter=1000)
log_reg3.fit(X_train_sm, y_train_sm)
```

```python
#Evaluating the model
y_pred_test_smote = log_reg3.predict(X_test_scaled_df)
print(classification_report(y_pred = y_pred_test_smote, y_true = y_test))
```

```python
#now with balanced data SMOTE, we check Does it improve the performance of our model? and the answer is it does improve
# compared to then unbalanced data, it goes from 60% up to 95% same as the score obtained with under and oversampling. So in conclusion
# balancing the data with any technique improves the prediction of frauds on 35% from the baseline model which is unbalanced.
```

- **4.** Run **Oversample** in order to balance our target variable and repeat the steps above, now with balanced data. Does it improve the performance of our model? 
- **5.** Now, run **Undersample** in order to balance our target variable and repeat the steps above (1-3), now with balanced data. Does it improve the performance of our model?
